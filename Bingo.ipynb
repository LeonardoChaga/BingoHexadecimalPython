```
import numpy
import pandas
```

```
def sortearSemRepeticao(INICIO = 16, FIM = 160, TAMANHO = 1):
  _dataset = numpy.zeros(TAMANHO, dtype=int)
  numero = numpy.random.randint(INICIO, FIM + 1, 1)                 
  _dataset[0] = numero
  l = 1       
  while (l < TAMANHO):
    numero = numpy.random.randint(INICIO, FIM + 1, 1)
    c = l - 1    
    haRepetidos = False   
    while c != -1:
      if _dataset[c] == numero:
        haRepetidos = True  
        break
      c -= 1
    if not haRepetidos:
      _dataset[l] = numero
      l += 1
  return _dataset
```

```
def sortearH():           
  Hex = str(hex(sortearSemRepeticao(INICIO = 16, FIM = 160, TAMANHO = 1)[0]))                   
  return Hex[2:]
```
